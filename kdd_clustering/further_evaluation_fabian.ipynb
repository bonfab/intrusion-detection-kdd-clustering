```python
def precision(TP, FP):
    return TP / (TP + FP)

def recall(TP, FN):
    return TP /(TP + FN)

def F1(precision, recall):
    return 2*precision*recall/(precision + recall)

def f1_cat(stats, cat):
    TP = 0
    FP = 0
    TN = 0
    FN = 0
    for i, row in stats.iterrows():
        #print(row)
        #print(row.idxmax(axis=1))
        #print(cat[i])
        if cat.name == row.idxmax(axis=1):
            TP += cat[i]
            FP += row.sum() - cat[i]
        elif not np.isnan(cat[i]):
            FN += cat[i]
            TN += row.sum() - cat[i]
    
    if TP == 0:
        return 0
    else:
        return F1(precision(TP, FP), recall(TP, FN))
            
```

```python
test = pd.read_csv('kddcup.testdata.unlabeled')
```

```python
def normalize(data):
    #data = data.drop('normal.', axis=1)
    def categorical_as_multicolumns(pd_data):
        for column_name in pd_data.columns:
            if pd_data[column_name].dtype.name == 'object':
                for category in np.unique(pd_data[column_name]):
                    pd_data[column_name + '-' + category] = pd_data[column_name] == category
                pd_data.drop(column_name, axis=1, inplace=True)
        return pd_data
    
    def normalize(pd_data):
        for column in pd_data.columns:
            std = pd_data[column].std()
            if np.allclose(std, 0):
                print(f"For column {column} too low std detected: {std}\nDropping column")
                pd_data.drop(column, axis=1, inplace=True)
            else:
                pd_data[column] = (pd_data[column] - pd_data[column].mean())/pd_data[column].std()
        return pd_data

    data = categorical_as_multicolumns(data)
    data = normalize(data)
    return data

test_norm = normalize(test)
```

```python
def project(test_data, data):
    global U
    global S
    global V
    U, S, V = np.linalg.svd(data.values.T, full_matrices = False)
    print(f'Calculated SVD. Mean of SVs: {S.mean()}')
    
    DIM_SUBSPACE = 20
    pcs = U[:,:DIM_SUBSPACE] @ np.diag(S[:DIM_SUBSPACE])
    pdata = test_data @ pcs
    return pdata

test_pdata = project(test_norm, normalize(pd.read_csv('kddcup.data').drop('normal.', axis=1)))
```

```python
from sklearn.cluster import KMeans
centroids = np.loadtxt(f"./centroids/{11}")
centroids.shape
```

```python
kmeans = KMeans(11, init=centroids, n_init=1, random_state=0).fit(test_pdata)
```

```python
np.linalg.norm((kmeans.cluster_centers_ - centroids), axis=1)
```
